# Análisis Exploratorio de Datos - Espirometrías

Este notebook contiene un análisis exploratorio completo de los datos de espirometrías.

## 1. Información Básica de los Datasets

```python
print("=== INFORMACIÓN BÁSICA DE LOS DATASETS ===")
print(f"\nDataset 1 (Df_all - Todos los intentos):")
print(f"Forma: {Df_all.shape}")
print(f"Columnas: {Df_all.shape[1]}")
print(f"Filas: {Df_all.shape[0]}")

print(f"\nDataset 2 (Df_full_variables - Mejor intento):")
print(f"Forma: {Df_full_variables.shape}")
print(f"Columnas: {Df_full_variables.shape[1]}")
print(f"Filas: {Df_full_variables.shape[0]}")

print(f"\n=== TIPOS DE DATOS ===")
print("\nDataset 1 - Tipos de datos:")
print(Df_all.dtypes.value_counts())

print("\nDataset 2 - Tipos de datos:")
print(Df_full_variables.dtypes.value_counts())
```

## 2. Análisis de Valores Faltantes

```python
def analyze_missing_values(df, dataset_name):
    print(f"\n=== VALORES FALTANTES - {dataset_name} ===")
    missing_data = df.isnull().sum()
    missing_percent = (missing_data / len(df)) * 100
    
    missing_df = pd.DataFrame({
        'Columna': missing_data.index,
        'Valores_Faltantes': missing_data.values,
        'Porcentaje': missing_percent.values
    })
    
    # Filtrar solo columnas con valores faltantes
    missing_df = missing_df[missing_df['Valores_Faltantes'] > 0]
    missing_df = missing_df.sort_values('Porcentaje', ascending=False)
    
    print(f"Total de columnas con valores faltantes: {len(missing_df)}")
    print(f"\nTop 10 columnas con más valores faltantes:")
    print(missing_df.head(10))
    
    return missing_df

# Analizar valores faltantes en ambos datasets
missing_df_all = analyze_missing_values(Df_all, "TODOS LOS INTENTOS")
missing_df_full = analyze_missing_values(Df_full_variables, "MEJOR INTENTO")
```

```python
# Visualización de valores faltantes
fig, axes = plt.subplots(2, 1, figsize=(15, 12))

# Dataset 1 - Top 15 columnas con más valores faltantes
if len(missing_df_all) > 0:
    top_missing_all = missing_df_all.head(15)
    axes[0].barh(top_missing_all['Columna'], top_missing_all['Porcentaje'])
    axes[0].set_title('Top 15 Columnas con Más Valores Faltantes - Todos los Intentos')
    axes[0].set_xlabel('Porcentaje de Valores Faltantes')
    axes[0].grid(True, alpha=0.3)

# Dataset 2 - Top 15 columnas con más valores faltantes
if len(missing_df_full) > 0:
    top_missing_full = missing_df_full.head(15)
    axes[1].barh(top_missing_full['Columna'], top_missing_full['Porcentaje'])
    axes[1].set_title('Top 15 Columnas con Más Valores Faltantes - Mejor Intento')
    axes[1].set_xlabel('Porcentaje de Valores Faltantes')
    axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 3. Análisis Demográfico

```python
# Usar el dataset de mejor intento para análisis demográfico
print("=== ANÁLISIS DEMOGRÁFICO ===")

# Análisis de género
print("\n1. Distribución por Género:")
gender_dist = Df_full_variables['Gender'].value_counts()
print(gender_dist)
print(f"Porcentaje: {(gender_dist/len(Df_full_variables)*100).round(2)}%")

# Análisis de raza
print("\n2. Distribución por Raza:")
race_dist = Df_full_variables['Race'].value_counts()
print(race_dist)
print(f"Porcentaje: {(race_dist/len(Df_full_variables)*100).round(2)}%")

# Análisis de edad
print("\n3. Estadísticas de Edad:")
if 'Age' in Df_full_variables.columns:
    age_stats = Df_full_variables['Age'].describe()
    print(age_stats)
else:
    print("Columna 'Age' no encontrada")

# Análisis de altura y peso
print("\n4. Estadísticas Antropométricas:")
if 'Height' in Df_full_variables.columns:
    print("Altura:")
    print(Df_full_variables['Height'].describe())
    
if 'Weight' in Df_full_variables.columns:
    print("\nPeso:")
    print(Df_full_variables['Weight'].describe())
```

```python
# Visualizaciones demográficas
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Distribución por género
gender_counts = Df_full_variables['Gender'].value_counts()
axes[0,0].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')
axes[0,0].set_title('Distribución por Género')

# Distribución por raza
race_counts = Df_full_variables['Race'].value_counts()
axes[0,1].pie(race_counts.values, labels=race_counts.index, autopct='%1.1f%%')
axes[0,1].set_title('Distribución por Raza')

# Histograma de edad (si existe)
if 'Age' in Df_full_variables.columns:
    Df_full_variables['Age'].hist(bins=20, ax=axes[1,0], edgecolor='black')
    axes[1,0].set_title('Distribución de Edad')
    axes[1,0].set_xlabel('Edad')
    axes[1,0].set_ylabel('Frecuencia')
    axes[1,0].grid(True, alpha=0.3)

# Scatter plot altura vs peso (si existen)
if 'Height' in Df_full_variables.columns and 'Weight' in Df_full_variables.columns:
    scatter = axes[1,1].scatter(Df_full_variables['Height'], Df_full_variables['Weight'], 
                               c=Df_full_variables['Gender'].map({'male': 'blue', 'female': 'red'}),
                               alpha=0.6)
    axes[1,1].set_title('Altura vs Peso por Género')
    axes[1,1].set_xlabel('Altura')
    axes[1,1].set_ylabel('Peso')
    axes[1,1].grid(True, alpha=0.3)
    axes[1,1].legend(['Masculino', 'Femenino'])

plt.tight_layout()
plt.show()
```

## 4. Análisis de Mediciones Espirométricas

```python
# Identificar variables espirométricas clave
spirometry_vars = ['FEV1', 'FVC', 'PEF', 'FEF25', 'FEF50', 'FEF75', 'MMEF', 'FEV6']

# Verificar qué variables están disponibles en el dataset
available_spiro_vars = [var for var in spirometry_vars if var in Df_full_variables.columns]
print(f"Variables espirométricas disponibles: {available_spiro_vars}")

# También buscar en el dataset de todos los intentos
available_spiro_vars_all = [var for var in spirometry_vars if var in Df_all.columns]
print(f"Variables espirométricas en dataset completo: {available_spiro_vars_all}")

# Estadísticas descriptivas de variables espirométricas
if available_spiro_vars:
    print("\n=== ESTADÍSTICAS DESCRIPTIVAS - VARIABLES ESPIROMÉTRICAS ===")
    spiro_stats = Df_full_variables[available_spiro_vars].describe()
    print(spiro_stats)
elif available_spiro_vars_all:
    print("\n=== ESTADÍSTICAS DESCRIPTIVAS - VARIABLES ESPIROMÉTRICAS (Dataset completo) ===")
    spiro_stats = Df_all[available_spiro_vars_all].describe()
    print(spiro_stats)
```

```python
# Distribuciones de variables espirométricas clave
dataset_to_use = Df_full_variables if available_spiro_vars else Df_all
vars_to_use = available_spiro_vars if available_spiro_vars else available_spiro_vars_all

def to_float_series(series):
    # Intenta convertir a float, reemplazando ',' por '.' si es necesario
    # Si ya es numérico, no hace nada
    if series.dtype == object:
        # Reemplaza ',' por '.' y fuerza a float, ignora errores
        return pd.to_numeric(series.str.replace(',', '.', regex=False), errors='coerce')
    else:
        return pd.to_numeric(series, errors='coerce')

if vars_to_use:
    n_vars = len(vars_to_use)
    n_cols = 3
    n_rows = (n_vars + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if n_vars > 1 else [axes]

    for i, var in enumerate(vars_to_use):
        if i < len(axes):
            # Filtrar valores no nulos y convertir a float
            data_raw = dataset_to_use[var].dropna()
            data = to_float_series(data_raw).dropna()
            if len(data) > 0:
                axes[i].hist(data, bins=30, edgecolor='black', alpha=0.7)
                axes[i].set_title(f'Distribución de {var}')
                axes[i].set_xlabel(var)
                axes[i].set_ylabel('Frecuencia')
                axes[i].grid(True, alpha=0.3)

                # Agregar estadísticas básicas
                mean_val = data.mean()
                median_val = data.median()
                axes[i].axvline(mean_val, color='red', linestyle='--', label=f'Media: {mean_val:.2f}')
                axes[i].axvline(median_val, color='green', linestyle='--', label=f'Mediana: {median_val:.2f}')
                axes[i].legend()
            else:
                axes[i].set_title(f'{var} (sin datos numéricos)')
                axes[i].set_visible(True)

    # Ocultar subplots vacíos
    for i in range(len(vars_to_use), len(axes)):
        axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron variables espirométricas estándar en los datasets")
```

## 5. Análisis de Correlaciones

```python
# Seleccionar variables numéricas para análisis de correlación
numeric_cols = dataset_to_use.select_dtypes(include=[np.number]).columns.tolist()

# Remover columnas con demasiados valores faltantes (>50%)
cols_to_analyze = []
for col in numeric_cols:
    missing_pct = dataset_to_use[col].isnull().sum() / len(dataset_to_use)
    if missing_pct < 0.5:  # Menos del 50% de valores faltantes
        cols_to_analyze.append(col)

print(f"Analizando correlaciones entre {len(cols_to_analyze)} variables numéricas")
print(f"Variables seleccionadas: {cols_to_analyze[:10]}...")  # Mostrar las primeras 10

if len(cols_to_analyze) > 1:
    # Calcular matriz de correlación
    correlation_matrix = dataset_to_use[cols_to_analyze].corr()
    
    # Encontrar las correlaciones más altas (excluyendo la diagonal)
    corr_pairs = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            corr_val = correlation_matrix.iloc[i, j]
            if not np.isnan(corr_val):
                corr_pairs.append({
                    'Variable1': correlation_matrix.columns[i],
                    'Variable2': correlation_matrix.columns[j],
                    'Correlacion': corr_val
                })
    
    # Ordenar por valor absoluto de correlación
    corr_df = pd.DataFrame(corr_pairs)
    corr_df['Correlacion_Abs'] = corr_df['Correlacion'].abs()
    corr_df = corr_df.sort_values('Correlacion_Abs', ascending=False)
    
    print("\n=== TOP 20 CORRELACIONES MÁS ALTAS ===")
    print(corr_df.head(20)[['Variable1', 'Variable2', 'Correlacion']])
```

```python
# Crear heatmap de correlación para variables más relevantes
if len(cols_to_analyze) > 1:
    # Seleccionar un subconjunto de variables para el heatmap (máximo 20)
    if len(cols_to_analyze) > 20:
        # Seleccionar variables con menos valores faltantes
        missing_counts = dataset_to_use[cols_to_analyze].isnull().sum()
        top_cols = missing_counts.sort_values().head(20).index.tolist()
    else:
        top_cols = cols_to_analyze
    
    plt.figure(figsize=(15, 12))
    correlation_subset = dataset_to_use[top_cols].corr()
    
    # Crear heatmap
    mask = np.triu(np.ones_like(correlation_subset, dtype=bool))
    sns.heatmap(correlation_subset, mask=mask, annot=True, cmap='viridis', center=0,
                square=True, linewidths=0.5, cbar_kws={"shrink": .8}, fmt='.2f')
    plt.title('Matriz de Correlación - Variables Numéricas Principales')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

## 6. Análisis de Calidad de Datos

```python
print("=== ANÁLISIS DE CALIDAD DE DATOS ===")

# Análisis de duplicados
print("\n1. Análisis de Duplicados:")
print(f"Dataset 1 - Filas duplicadas: {Df_all.duplicated().sum()}")
print(f"Dataset 2 - Filas duplicadas: {Df_full_variables.duplicated().sum()}")

# Análisis de IDs únicos
print("\n2. Análisis de IDs:")
if 'Id' in Df_all.columns:
    print(f"Dataset 1 - IDs únicos: {Df_all['Id'].nunique()} de {len(Df_all)} registros")
if 'Id Paciente' in Df_full_variables.columns:
    print(f"Dataset 2 - IDs únicos: {Df_full_variables['Id Paciente'].nunique()} de {len(Df_full_variables)} registros")

# Análisis de fechas
print("\n3. Análisis de Fechas:")
date_columns = ['Visit Date', 'Test Date', 'Date_of_Birth', 'Date First Test', 'Date Last Test']
for col in date_columns:
    if col in dataset_to_use.columns:
        non_null_dates = dataset_to_use[col].dropna()
        if len(non_null_dates) > 0:
            print(f"{col}: {len(non_null_dates)} valores válidos")

# Análisis de valores extremos en variables numéricas clave
print("\n4. Análisis de Valores Extremos:")
for var in vars_to_use[:5]:  # Analizar las primeras 5 variables espirométricas
    if var in dataset_to_use.columns:
        data = dataset_to_use[var].dropna()
        # Intentar convertir a numérico, forzar errores a NaN y eliminar los NaN resultantes
        data_numeric = pd.to_numeric(data, errors='coerce').dropna()
        if len(data_numeric) > 0:
            Q1 = data_numeric.quantile(0.25)
            Q3 = data_numeric.quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = data_numeric[(data_numeric < lower_bound) | (data_numeric > upper_bound)]
            print(f"{var}: {len(outliers)} valores extremos ({len(outliers)/len(data_numeric)*100:.1f}%)")
        else:
            print(f"{var}: No hay datos numéricos válidos para analizar valores extremos.")
```

```python
# Visualización de valores extremos con boxplots
if vars_to_use:
    n_vars_plot = min(6, len(vars_to_use))  # Máximo 6 variables
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    axes = axes.flatten()
    
    for i, var in enumerate(vars_to_use[:n_vars_plot]):
        data = dataset_to_use[var].dropna()
        # Convertir a numérico, forzar errores a NaN y eliminar los NaN resultantes
        data_numeric = pd.to_numeric(data, errors='coerce').dropna()
        if len(data_numeric) > 0:
            axes[i].boxplot(data_numeric)
            axes[i].set_title(f'Boxplot - {var}')
            axes[i].set_ylabel(var)
            axes[i].grid(True, alpha=0.3)
        else:
            axes[i].set_visible(False)
    
    # Ocultar subplots vacíos
    for i in range(n_vars_plot, len(axes)):
        axes[i].set_visible(False)
    
    plt.tight_layout()
    plt.show()
```
